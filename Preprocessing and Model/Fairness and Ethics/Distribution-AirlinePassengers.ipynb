```python
import pandas as pd
import matplotlib.pyplot as plt

# read the csv file into a pandas dataframe
df = pd.read_csv("test.csv")

# calculate the distribution of the Gender variable
gender_distribution = df['Gender'].value_counts(normalize=True)

# plot the distribution in a bar chart
fig, ax = plt.subplots(figsize=(6,4))
gender_distribution.plot(kind='bar', ax=ax, width=0.9, rot=0)

# set y-axis to whole numbers
ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))

# display the value for each bar graph
for i, v in enumerate(gender_distribution):
    ax.text(i - 0.1, v + 0.01, str(round(v * len(df))), fontsize=8)
    ax.text(i - 0.1, v - 0.05, str(round(v * 100, 1))+'%', fontsize=8)

# set chart title and axis labels
ax.set_title('Distribution of Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')

# set y-axis range and tick frequency
```

```python
import pandas as pd
import matplotlib.pyplot as plt

X_val = pd.read_csv("test.csv")

young_mask = X_val['Age'] <= 25
middle_age_mask = (X_val['Age'] > 25) & (X_val['Age'] <= 50)
old_mask = X_val['Age'] > 50

young_count = young_mask.sum()
middle_age_count = middle_age_mask.sum()
old_count = old_mask.sum()

total_count = len(X_val)
young_percent = round(young_count / total_count * 100, 1)
middle_age_percent = round(middle_age_count / total_count * 100, 1)
old_percent = round(old_count / total_count * 100, 1)


fig, ax = plt.subplots()
ax.bar(['Young', 'Middle-aged', 'Old'], [young_percent, middle_age_percent, old_percent], width=0.8)


for i, v in enumerate([young_percent, middle_age_percent, old_percent]):
    ax.text(i - 0.05, v + 0.5, str(v) + '%', fontsize=10, ha='center')


ax.set_title('Distribution of Age')
ax.set_xlabel('Age Group')
ax.set_ylabel('Percentage')

plt.show()
```
