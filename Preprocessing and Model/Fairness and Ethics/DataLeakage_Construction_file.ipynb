# Data Leakage Construction File - File was created in order to generate PII Information for Datset

```python
import pandas as pd
import numpy as np
import string
import random
import requests

first_names = []
response = requests.get('https://www.mit.edu/~ecprice/wordlist.10000')
for line in response.iter_lines():
    first_names.append(line.decode('utf-8'))

last_names = []
response = requests.get('https://www.mit.edu/~ecprice/wordlist.10000')
for line in response.iter_lines():
    last_names.append(line.decode('utf-8'))


def generate_credit_card_number():
    # Generate a random 15-digit number
    digits = [random.randint(0, 9) for _ in range(15)]
    for i in range(14, -1, -2):
        digits[i] *= 2
        if digits[i] > 9:
            digits[i] -= 9
    check_digit = (10 - sum(digits) % 10) % 10
    return f"{digits[0:4]}-{digits[5:9]}-{digits[10:14]}-{check_digit}"


train_airplane = pd.read_csv('/Users/pedrojosetrujillomejia/Desktop/airplane_test_processed_normalized_date.csv')
test_airplane = pd.read_csv('/Users/pedrojosetrujillomejia/Desktop/airplane_train_processed_date.csv')

combined_dataframe = pd.concat([train_airplane, test_airplane], ignore_index=True)

num_rows = combined_dataframe.shape[0]

first_names = np.random.choice(first_names, num_rows)
last_names = np.random.choice(last_names, num_rows)
credit_cards = np.array([generate_credit_card_number() for i in range(num_rows)])

combined_dataframe['First Name'] = first_names
combined_dataframe['Last Name'] = last_names
combined_dataframe['Credit Card Information'] = credit_cards
```
