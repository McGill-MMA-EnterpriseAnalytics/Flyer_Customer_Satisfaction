Although our chosen dataset did not include a date as one of its predictors, we found it advisable to include one for our current use case. Incorporating a date allows us to perform tasks such as data versioning, data drift checks, etc. To address this, we artificially generated dates starting from January 1, 2017, to December 31, 2022.

```python
import pandas as pd
import random
import datetime

# Create a Pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/Enterprise Data Science Group Project 2/test.csv')
```

```python
# Generate random dates and add to DataFrame
start_date = datetime.date(2017, 1, 1)
end_date = datetime.date(2022, 12, 31)
date_list = [start_date + datetime.timedelta(days=random.randint(0, (end_date - start_date).days)) for _ in range(len(df))]  
df['Date'] = date_list
```

```python
df.to_csv("test_date.csv",index=False)
```
